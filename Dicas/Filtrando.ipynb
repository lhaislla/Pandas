## Filtrando e Exportando para CSV os dados do ENEM por Estado Brasileiro

Importando o pacote <strong>Pandas</strong> único que será utilizado nessa aula:

<em>Documentação do Pandas</em>: https://pandas.pydata.org/pandas-docs/stable/index.html

Link para os microdados do enem 2019: http://inep.gov.br/microdados

```python
import pandas as pd
import zipfile
```

```python
with zipfile.ZipFile('../Data/Formatos_diversos/microdados_enem_2019.zip') as z:
    print(*z.namelist(),sep="\n")
```

```python
with zipfile.ZipFile('../Data/microdados_enem_2019.zip') as z:
    with z.open('DADOS/MICRODADOS_ENEM_2019.csv') as f:
        %time microdados_enem = pd.read_csv(f,sep=';',encoding='ISO-8859-1')
```

```python
lista_ufs = sorted(microdados_enem['SG_UF_RESIDENCIA'].unique())
```

```python
len(lista_ufs)
```

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html

```python
for uf in lista_ufs:
    caminho_do_arquivo = "../dados/microdados_enem_{}_2019.csv".format(uf)
    microdados_enem[microdados_enem['SG_UF_RESIDENCIA']== uf].to_csv(caminho_do_arquivo, index=False)
```
