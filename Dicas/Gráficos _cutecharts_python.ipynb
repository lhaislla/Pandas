```python
import pandas as pd
import cutecharts.charts as ctc
```

```python
microdados_enem_rj = pd.read_csv('../Data/microdados_enem_ufs_2019/microdados_enem_RJ_2019.csv', index_col=0)
```

```python
chart = ctc.Bar('Médias Notas RJ',width='500px',height='400px')
chart.set_options(
 labels=['C. Natureza', 'C. Humanas', 'Linguagens', 'Matemática', 'Redação'],
 x_label="Provas",
 y_label="Média por Prova" ,
 colors=['#D9F1BB','#F7B7A3','#EA5F89','#9B3192', '#FFFF77']
 )
chart.add_series("Média",list(microdados_enem_rj[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean()))
chart.render_notebook()
```

```python
municipio = "Angra dos Reis"

radar_chart = ctc.Radar('Médias Notas RJ',width='600px',height='600px')
radar_chart.set_options(
 labels=['C. Natureza', 'C. Humanas', 'Linguagens', 'Matemática'],
 is_show_legend=True, #by default, it is true. You can turn it off.
 legend_pos='upRight'  #location of the legend
 )

radar_chart.add_series("Média dos Municípios RJ",list(microdados_enem_rj[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT','NU_NOTA_REDACAO']].mean()))
radar_chart.add_series("Média do Município {}".format(municipio),
                       list(microdados_enem_rj.groupby(['NO_MUNICIPIO_RESIDENCIA'])['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT','NU_NOTA_REDACAO'].mean().loc[municipio]))

radar_chart.render_notebook()
```
