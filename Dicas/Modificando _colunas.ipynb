## Configurando a apresentação de um DataFrame

link para documentação: https://pandas.pydata.org/pandas-docs/stable/user_guide/options.html

```python
import pandas as pd
```

Link para o conjunto de dados: http://www.dados.gov.br/dataset/precos-de-medicamentos-governo

```python
FILE_PATH = '../Data/TA_PRECO_MEDICAMENTO_GOV.csv'
```

```python
meds = pd.read_csv(FILE_PATH, encoding="ISO-8859-1", sep=';')
```

```python
meds.head()
```

```python
pd.get_option('max_colwidth')
```

```python
meds['APRESENTAÇÃO'].str.len().max()
```

```python
pd.set_option('max_colwidth',170)
```

```python
meds.head()
```

```python
pd.get_option('max_columns')
```

```python
pd.set_option('max_columns', 50)
```
